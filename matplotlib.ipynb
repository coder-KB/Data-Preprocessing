##  Simple Line Plot using pyplot

```python
import matplotlib.pyplot as plt #importing pyplot
%matplotlib inline
plt.plot([1, 2, 3, 4],[2,1,4,5]) #these arrays acts as co-ordinates for the plot
plt.show()
```

```python
plt.plot([1, 2, 3, 4])#only one array as parameter
plt.ylabel("Y axis")
plt.xlabel("x Axis")
plt.title("Demo line plot")
plt.show()
```

# Scatter Plots

```python
x=[1, 2, 3,4]
y=[5,6,7,8]
plt.scatter(x, y, marker='s');#s gives square ,o gives circle, p gives pentagon
```

```python
import numpy as np
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o', color='red');
```
